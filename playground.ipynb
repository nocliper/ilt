```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.cm import gnuplot, jet

import sys
sys.path.append('functions/')

%matplotlib notebook
```

```python
df = pd.read_csv('V735_N2_4e13GeVpr_150_1_reSpect(1e-2).LDLTS', delimiter = '\t')
columns = np.asarray(df.columns)
columns[0] = 'Temperature'
df.columns = columns
df
```

```python
fig = plt.figure(figsize = (9,4))
ax  = fig.add_subplot()

color = np.linspace(0, 1, len(columns) - 1)
color = gnuplot(color)

for i, c in enumerate(columns[1:]):
    if i%10 == 0:
        ax.plot(df['Temperature'], df[c]/max(df[c]), c = color[i],
                label = r'$\tau$ = %.2f s'%(float(c)), lw = 3)
ax.legend()
plt.tight_layout()
```

```python
from L1_FISTA import l1_fista
```

```python
t = np.linspace(150, 150*200, 200)*1e-3
f = -np.exp(-0.5*t) + np.exp(-5*t) + np.exp(-0.05*t)

es = [0.05, 0.5, 5]

z, F, f_hat = l1_fista(t, f, [1e-3, 1e+3], 500, 1, 
                       iterations = 50000, penalty = 'l11')

fig = plt.figure(figsize = (9,4))
ax  = fig.add_subplot()
ax.set_xscale('log')

for es_ in es:
    ax.axvline(1/es_)

ax.plot(t, f)
ax.plot(t, f_hat, '--')
ax.plot(1/z, F)
```
