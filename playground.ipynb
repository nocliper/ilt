```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
df = pd.read_csv('V735_N2_4e13GeVpr_150_1_reSpect(1e-2).LDLTS', delimiter = '\t')
columns = np.asarray(df.columns)
columns[0] = 'Temperature'
df.columns = columns
df
```

```python
fig = plt.figure()
ax  = fig.add_subplot()

for i, c in enumerate(columns[1:-20]):
    if i%10 == 0:
        ax.plot(df['Temperature'], df[c]/max(df[c]), label = r'$\tau$ = %.2f s'%(float(c)))
ax.legend()
plt.tight_layout()
```
