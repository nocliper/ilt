```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
```

```python
T = np.linspace(130, 400, 400)
t = np.arange(0.150, 15, 0.150)
```

```python
Ea  = [0.74,   0.75] #eV
sig = [3E-15, 5E-14] #cm**2
Nt  = [4E16,   4E16] #cm-3
Nd  = 1E16

e = []

m   = 0.28*9.109E-31 #kg
k   = 1.380649E-23 #J/K
h   = 6.62607015E-34 #J*s
Nc  = 2*(2*np.pi*m*k*T/h**2)**(1.5) #m**-3
Nc  = Nc*10**-6 #cm-3
v   = np.sqrt(3*k*T/m) #m/s
v   = v*100 #cm/s
k   = 8.617333262E-5 #eV/K

for i in range(len(Ea)):
    temp = sig[i]*v*Nc*np.exp(-Ea[i]/(k*T)) #
    e.append(temp)
```

```python
C = []
C0  = 20*np.log(T) + 3*np.cos(T/30) + 600 # Стационарная емкость
dC0 =np.gradient(C0)/np.gradient(T)
dC0 = np.append(dC0, dC0[-1])

for i in range(len(T)):
    temp = C0[i]
    for j in range(len(Ea)):
        temp = temp - 100*Nt[j]/2/Nd*np.exp(-e[j][i]*t)
        temp = temp + np.random.normal(0, 1, len(t))*9e-2
        temp = temp + dC0[i]/60*1.5*t #
    C.append(temp)
```

```python
for i in C:
    plt.plot(t, i)
```

```python
Table = []

for i in range(len(T)):
    Table.append([T[i]] + (C[i]).tolist())
```

```python
np.savetxt('test_150.DLTS', Table, delimiter='\t', fmt = '%4E')
```
