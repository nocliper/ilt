```python
import matplotlib.pyplot as plt
import numpy as np

import sys
sys.path.append('functions/')
from ilt import SVD

#from read_file import read_file
#from interface import interface

%matplotlib inline
plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"
```

```python
def adv_average(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
t = np.arange(0.15, 15, 0.15)
C = 2*np.exp(-1*t) + 3*np.exp(-8*t)+3
C = C + 0.0001*t
C = C + np.random.normal(0, 1, len(t))*5e-2
p = np.zeros(len(t))
p[int(len(t)/2)] = C[int(len(t)/2)]*0.3
p[int(len(t)/10)] = -C[int(len(t)/10)]*0.2
C = C + p
plt.plot(t, C)
#plt.plot(t, adv_average(C, 1))
```

```python
def plot(t, C):
    
    fig = plt.figure()
    ax  = fig.add_subplot(111)
    ax.set_xscale('log')

    ax.plot(t, C)
    z, f, C_r = SVD(t, C - C[-1], [1e-2, 1e2], 100, 0.1)
    ax.axvline(x=1/1, ls = ':')
    ax.axvline(x=1/8, ls = ':')
    ax.plot(1/z, f)
```

```python
plot(t, C)
```

```python
plot(t, C)
plot(t, adv_average(C, 7))
```
